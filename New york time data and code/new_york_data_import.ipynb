```python

from bs4 import BeautifulSoup
import requests
import json

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?q=blockchain&begin_date=20180210&end_date=20190119&page=1&sort=oldest&api-key=XSMLqsXNgMihVRmUtTvry3rvjRKL37XT"
resp = requests.get(url)
print(resp)
data = json.loads(resp.text)
print(data)
```

```python
for x in range(0, 10):
    new_url = data["response"]["docs"][x]["web_url"]
    print(data["response"]["docs"][x]["web_url"])
    new = requests.get(new_url)
    soup = BeautifulSoup(new.text,'html.parser')
    results = soup.find_all('p',attrs={'class':'css-1ygdjhk evys1bk0'})
    filename = "block" + str(10 + x ) + ".txt"
    text_file = open(filename,"w")
    for i in range(0,len(results)-1) :
        text_file.write(results[i].text)
```
